```python
%matplotlib notebook
import nibabel as nib
import matplotlib.pyplot as plt


image_t1 = nib.load("small_data/small/t1/IXI102-HH-1416-T1_fcm.nii.gz").get_fdata()
image_t2 = nib.load("small_data/small/t2/IXI102-HH-1416-T2_reg_fcm.nii.gz").get_fdata()
```
